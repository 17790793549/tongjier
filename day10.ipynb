# 会话维持

会话cookie误区

```python
import requests
requests.get('http://httpbin.org/cookies/set/number/123456')
response=requests.get('http://httpbin.org/cookies')
print(response.text)
```

# 产生无法获取到cookies的原因是由于上面两次get实质上是类似于两个浏览器的访问，故第二个浏览器无法获取到第一个浏览器所存的cookies

```python
import requests
s = requests.Session()#通过声明Session告诉两次用的是一个浏览器
s.get('http://httpbin.org/cookies/set/number/123456')
response=s.get('http://httpbin.org/cookies')
print(response.text)
```

# 设置超时

timeout=1 设置时间为一秒，即大于一秒为超时

```python
import requests
response = requests.get('http://www.baidu.com',timeout = 1)
print(response.status_code)
```

# 错误处理

ReadTimeout 连接超时
ConnectionError  连接错误
RequestException 请求错误

```python
import requests
try:
    response = requests.get('http://www.baidu.com',timeout=0.01)
    print(response.status_code)
except Exception as e:
    print(e)
```

# Example

定义一个函数的形式
抓取http://www.17k.com/的任意两个网页，存入本地

```python
def liuyu():
    import requests
    with open('C:/Users/刘钰/Desktop/asd.txt','w',encoding='utf-8') as f:
        for i in range(2):
            aaa =['http://www.17k.com/book/2801193.html','http://www.17k.com/']
            try:
                response = requests.get(aaa[i],timeout=1)
                f.write(response.text)
            except Exception as e:
                print(e)
    f.close()
```
