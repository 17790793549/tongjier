XPATH语法

一层一层往下挖模式

- '/'	从根节点选取
  '//'	从匹配选择的当前节点选择文档中的节点，而不考虑它们的位置 **
   '.'	选取当前节点 **
   '..'	选取当前节点的父节点
- '@'   提取属性和控制属性 ** 
- '[]'  属性容器
  '*'	匹配任何元素节点

- text() 以文本的方式返回提取的内容 **


```python
tree.xpath('//bookstore/book/title[@lang="engg"]/text()')
```
